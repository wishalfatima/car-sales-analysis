```python
import pandas as pd

# Load the datasets
car_sales_path = 'Car_sales.csv'
gcar_data_path = 'gcar_data.csv'

# Reading the data into pandas dataframes
car_sales_df = pd.read_csv(car_sales_path)
gcar_data_df = pd.read_csv(gcar_data_path)

# Display the first few rows of each dataframe to understand their structure
(car_sales_df.head(), gcar_data_df.head())
```

```python
# Data Cleaning: Handling missing values and removing any potential duplicate rows for both datasets.

# Check for missing values in each dataset
missing_values_car_sales = car_sales_df.isnull().sum()
missing_values_gcar_data = gcar_data_df.isnull().sum()

# Display the missing values count for each column in both datasets
(missing_values_car_sales, missing_values_gcar_data)
```

```python
# Remove 'Unnamed: 0' from gcar_data_df
import pandas as pd

# Remove 'Unnamed: 0' from gcar_data_df exists
if 'Unnamed: 0' in gcar_data_df.columns:
    gcar_data_df.drop('Unnamed: 0', axis=1, inplace=True)

# Define numerical columns for car_sales_df
numerical_columns_car_sales = [
    '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower', 
    'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity', 'Fuel_efficiency', 
    'Power_perf_factor'
]

# Fill missing numerical values in 'Car_sales.csv' with median
for col in numerical_columns_car_sales:
    median_value = car_sales_df[col].median(skipna=True)  # Ensure to skip NA values for median calculation
    car_sales_df[col].fillna(median_value, inplace=True)

# Define numerical columns for gcar_data_df
numerical_columns_gcar_data = ['power_kw', 'power_ps', 'mileage_in_km']

# Fill missing numerical values in 'gcar_data.csv' with median
for col in numerical_columns_gcar_data:
    median_value = gcar_data_df[col].dropna().median()  # Exclude NA for accurate median calculation
    gcar_data_df[col].fillna(median_value, inplace=True)

# Fill missing 'color' in 'gcar_data.csv' with the mode (most common value)
if gcar_data_df['color'].isna().any():  # Check if there's any NA in 'color' before trying to fill it
    gcar_data_df['color'].fillna(gcar_data_df['color'].mode()[0], inplace=True)

# Check if the data is cleaned properly by printing the sum of null values in each DataFrame
print(car_sales_df.isnull().sum(), "\n")
print(gcar_data_df.isnull().sum())

```

```python
# Convert 'power_kw' and 'power_ps' to numeric values in 'gcar_data.csv' by removing any non-numeric characters and converting
import numpy as np

# Define a function to convert strings with non-numeric characters to float
def convert_to_float(s):
    try:
        # Attempt to directly convert to float
        return float(s)
    except:
        # If conversion fails, try removing commas and converting
        try:
            return float(s.replace(',', '.'))
        except:
            # If still fails, convert to NaN
            return np.nan

# Apply the conversion to 'power_kw' and 'power_ps'
gcar_data_df['power_kw'] = gcar_data_df['power_kw'].apply(convert_to_float)
gcar_data_df['power_ps'] = gcar_data_df['power_ps'].apply(convert_to_float)

# Now attempt to fill missing values again with the median for these corrected columns
for col in numerical_columns_gcar_data:
    median_value = gcar_data_df[col].median()  # Recalculate median without errors
    gcar_data_df[col].fillna(median_value, inplace=True)

# Verify the correction by checking data types and missing values count
(gcar_data_df.dtypes, gcar_data_df.isnull().sum())
```

```python
# Fill missing 'color' with the most common value
gcar_data_df['color'].fillna(gcar_data_df['color'].mode()[0], inplace=True)

# Convert 'price_in_euro' to numeric, assuming the values might have commas or other non-numeric characters
gcar_data_df['price_in_euro'] = gcar_data_df['price_in_euro'].apply(lambda x: pd.to_numeric(x.replace(',', ''), errors='coerce'))

# Fill any new missing values in 'price_in_euro' that resulted from conversion errors with the median
median_price = gcar_data_df['price_in_euro'].median()
gcar_data_df['price_in_euro'].fillna(median_price, inplace=True)

# Display data type for 'price_in_euro' and check for remaining missing values across the dataset
(gcar_data_df.dtypes['price_in_euro'], gcar_data_df.isnull().sum())

```

```python
# Fill the single missing 'registration_date' with the most common value in 'gcar_data.csv'
gcar_data_df['registration_date'].fillna(gcar_data_df['registration_date'].mode()[0], inplace=True)

# Check the updated count of missing values across all columns to ensure all are handled
gcar_data_df.isnull().sum()

```

```python
# Redo the merge with an inner join to eliminate non-matching entries and minimize missing data
merged_df_inner = pd.merge(car_sales_df, gcar_data_df, left_on=['Manufacturer', 'Model'], right_on=['brand', 'model'], how='inner')

# Recreate Dimension_Car from the inner merged dataset with less missing data
dimension_car_inner = merged_df_inner[['Manufacturer', 'Model', 'color', 'Vehicle_type', 'price_in_euro', 'power_kw']].copy()
dimension_car_inner.drop_duplicates(inplace=True)

# Create other dimension tables based on the schema
# Dimension_Date
dimension_date = merged_df_inner[['Latest_Launch']].copy()
dimension_date['Year'] = pd.to_datetime(dimension_date['Latest_Launch']).dt.year
dimension_date['Month'] = pd.to_datetime(dimension_date['Latest_Launch']).dt.month
dimension_date['Quarter'] = pd.to_datetime(dimension_date['Latest_Launch']).dt.quarter
dimension_date.drop_duplicates(inplace=True)

# Dimension_Fuel
dimension_fuel = merged_df_inner[['fuel_type', 'Fuel_capacity', 'Fuel_efficiency']].copy()
dimension_fuel.drop_duplicates(inplace=True)

# Display the first few rows of each dimension table and the shape to verify
(dimension_car_inner.head(), dimension_car_inner.shape, dimension_date.head(), dimension_date.shape, dimension_fuel.head(), dimension_fuel.shape)

```

```python
# Display unique Manufacturer and Model values from both datasets to understand discrepancies
unique_manufacturers_car_sales = car_sales_df['Manufacturer'].unique()
unique_models_car_sales = car_sales_df['Model'].unique()
unique_brands_gcar_data = gcar_data_df['brand'].unique()
unique_models_gcar_data = gcar_data_df['model'].unique()

# Display a sample of unique values from each dataset for comparison
(unique_manufacturers_car_sales[:10], unique_models_car_sales[:10], unique_brands_gcar_data[:10], unique_models_gcar_data[:10])
```

```python
import re

# Function to simplify model names to match the simpler format of car_sales.csv
def simplify_model_name(model_name):
    # Remove any additional descriptors after the main model name (e.g., 'ford kuga' -> 'kuga')
    simplified_name = re.split('\s+', model_name.strip())[1] if len(re.split('\s+', model_name.strip())) > 1 else model_name
    return simplified_name

# Apply the simplification to the 'model' column in gcar_data_df
gcar_data_df['simplified_model'] = gcar_data_df['model'].apply(simplify_model_name)

# Retry merging using simplified model names
merged_df_simplified = pd.merge(car_sales_df, gcar_data_df, left_on=['Manufacturer', 'Model'], right_on=['brand', 'simplified_model'], how='inner')

# Recreate Dimension_Car with the new merge
dimension_car_simplified = merged_df_simplified[['Manufacturer', 'Model', 'color', 'Vehicle_type', 'price_in_euro', 'power_kw']].copy()
dimension_car_simplified.drop_duplicates(inplace=True)

# Display the first few rows of the new dimension table and the shape to verify success of the merge
(dimension_car_simplified.head(), dimension_car_simplified.shape)


#print(gcar_data_df.columns)

#print(car_sales_df.columns)

print (dimension_car_simplified.columns)
#gcar_data_df.to_csv('dimension_car_simplified.csv', index=False)
gcar_data_df.to_csv('gcar_data_df.csv', index=False)
#car_sales_df.to_csv('car_sales_df.csv', index=False)
```
